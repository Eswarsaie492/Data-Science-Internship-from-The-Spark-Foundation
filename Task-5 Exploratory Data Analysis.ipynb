```python
#Outlier Treatment
#Z-Score Outlier Treatment

z = np.abs(stats.zscore(df['Sales']))
z
```

```python
threshold = 3
np.where(z > 3)
```

```python
print(z[27])
```

```python
df_o = df.Sales[(z < 3)]
df_o.shape
```

```python
sns.distplot(df_o);
```

```python
print("Skewness after zscore outlier treatment : %f" % df_o.skew())
print("Kurtosis after zscore outlier treatment : %f" % df_o.kurt())
```

```python
#IQR Outlier Treatment

Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
print('Value of 1st quartile',Q1)
print('Value of 3rd quartile',Q3)
print('Interquartile Range',IQR)
```

```python
df_out = df[~((df['Sales']< (Q1 - 1.5 * IQR)) |(df['Sales'] > (Q3 + 1.5 * IQR)))]
df_out.shape
```

```python
sns.distplot(df_out['Sales']);
```

```python
print("Skewness after IQR outlier treatment : %f" % df_out['Sales'].skew())
print("Kurtosis after IQR outlier treatment : %f" % df_out['Sales'].kurt())
```

```python
print("Skewness")
print("Skewness with outliers: %f" % df['Sales'].skew())
print("Skewness after zscore outlier treatment : %f" % df_o.skew())
print("Skewness after IQR outlier treatment : %f" % df_out['Sales'].skew())
print("\nKurtosis")
print("Kurtosis with outliers: %f" % df['Sales'].kurt())
print("Kurtosis after zscore outlier treatment : %f" % df_o.kurt())
print("Kurtosis after IQR outlier treatment : %f" % df_out['Sales'].kurt())
```

```python
sns.jointplot(x='Sales',y='Profit',data=df_out)
plt.show()
```
